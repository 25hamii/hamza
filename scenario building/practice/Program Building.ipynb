```python
# Define initial empty grocery list and purchased items set
grocery_list = []
purchased_items = set()

# Add items to the grocery list
print("Adding items to the grocery list:")
user_input = input("Enter grocery items separated by comma: ")
grocery_list.extend(user_input.split(','))
print("Items added to the grocery list.")

# Display the current list of items
print("\nCurrent Grocery List:")
for item in grocery_list:
    print(item)

# Remove items from the grocery list
print("\nRemoving items from the grocery list:")
user_input = input("Enter items to remove separated by comma: ")
items_to_remove = user_input.split(',')
for item in items_to_remove:
    if item in grocery_list:
        grocery_list.remove(item)
        if item in purchased_items:
            purchased_items.remove(item)
        print(f"'{item}' removed from the grocery list.")
    else:
        print(f"'{item}' is not in the list.")

# Display the current list of items after removal
print("\nModified List:")
for item in grocery_list:
    print(item)

# Mark items as purchased
print("\nMarking items as purchased:")
user_input = input("Enter items to mark as purchased separated by comma: ")
items_to_mark = user_input.split(',')
for item in items_to_mark:
    if item in grocery_list:
        purchased_items.add(item)
        print(f"'{item}' marked as purchased.")
    else:
        print(f"'{item}' is not in the list.")

# Display the current list with updated item statuses
print("\nUpdated List with Purchased Status:")
for item in grocery_list:
    status = "Purchased" if item in purchased_items else "Not Purchased"
    print(f"{item} - {status}")

# Display a summary of purchased items
print("\nSummary of Purchased Items:")
for item in purchased_items:
    print(item)
```

```python
# Scenario: Calculator Program

#You are tasked with creating a basic calculator program that can perform addition, subtraction,
#multiplication,
#and division operations on two numbers.

while True:
    first_num = int(input("Please enter the first number: "))
    second_num = int(input("Please enter the second number: "))

    operation_choice = input("choose an operation (+, -, *, /): ")


    if operation_choice == '+':
        result = first_num + second_num 
    elif operation_choice == '-':
        result = first_num - second_num
    elif operation_choice == '*':
        result = first_num * second_num
    elif operation_choice == '/':
        if second_num == 0:
            print("Error division by zero")
            break
        
        else:
            result = first_num / second_num

    else:
        print("invalid operation")
        continue
    

    print("result", result)
    
    perform_another_calculation = input("Do you want to perform another calculation? (yes/no): ")    
    if perform_another_calculation == "no":
        print("Thankyou for using the calculator program")
        break
```

```python
print("Welcome to the Grocery Shopping list program")

grocery_list = []

while True:
    print("\nMenu: ")
    print("1. Add an item to the list: ")
    print("2. Remove an item from the list: ")
    print("3. View the current items on a list: ")
    print("4. Quit the program: ")
    
    choice = input("Enter your choice (1/2/3/4): ")
    if choice == '1':
        item = input("enter the item to add to the list: ")
        grocery_list.append(item)
        print(f"{item} has been added to the list.")

    elif choice == '2':
        item = input("enter the item to remove from the list: ")
        if item in grocery_list:
            grocery_list.remove(item)
            print(f"{item} has been removed from the list.")
        else:
            print("Item not found in the list.")
    
    elif choice == '3':
        item = input("view the current items of list")
        for item in grocery_list:
            print(item)
    
    elif choice == '4':
        print("Goodbye!")
        break

    else:
        print("Invalid choice please enter 1,2,3 or 4.")
```

```python
print("Welcome to Student Grade Tracker")

student_grade_tracker = {}

while True:
    print("\nMenu:")
    print("1. Add a new student")
    print("2. Enter grades for a student")
    print("3. Calculate average grade for a student")
    print("4. Display average grades for all students")
    print("5. Quit the program")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        new_student_name = input("Enter the student's name: ")
        student_grade_tracker[new_student_name] = []
        print(f"Student {new_student_name} added successfully!")
        
    elif choice == '2':
        student_name = input("Enter the student's name: ")
        grades_input = input("Enter the grades (separated by commas): ").split(',')
        grades = [int(grade) for grade in grades_input]
        student_grade_tracker[student_name] = grades
        print(f"Grades added successfully for student {student_name}!")
        
    elif choice == '3':
        student_name = input("Enter the student's name: ")
        grades = student_grade_tracker.get(student_name, [])
        if grades:
            average_grade = sum(grades) / len(grades)
            print(f"Average grade for student {student_name}: {average_grade}")
        else:
            print(f"No grades found for student {student_name}.")
        
    elif choice == '4':
        print("Average grades for all students:")
        for student, grades in student_grade_tracker.items():
            average_grade = sum(grades) / len(grades)
            print(f"{student}: {average_grade}")
            
    elif choice == '5':
        print("Goodbye! Thank you for using the Student Grade Tracker!")
        break
    
    else:
        print("Invalid choice. Please enter a number from 1 to 5.")
```

```python
# Coin toss game

import random

print("Welcome to the Coin Toss Game!")
print("I will flip a coin, and you will try to guess the outcome.")

score = 0

while True:
    guess = input("Enter your guess (heads or tails): ")
    coin_toss_result = random.randint(0,1)
    
    if guess == 'heads' and coin_toss_result == 0:
        print("Congratulations, You guessed correctly")
        score += 1
        
    elif guess == 'tails' and coin_toss_result == 1:
        print("Congratulations, You guessed correctly")
        score += 1
    
    else:
        print("Sorry! You guessed incorrectly.")
        
    play_again = input("Do you want to play again? (yes/no): ")
    
    if play_again != 'yes':
        print("Thanks for playing! your score is:", score)
        break
    else:
        print("\nLet's play again\n")
```
